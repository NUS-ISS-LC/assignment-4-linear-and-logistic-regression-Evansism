## Logistic Regression Workshop

```python
import numpy as np
import pandas as pd

df = pd.read_csv('iris-data-clean.csv')
df['class'] = df['class'].replace({
    'Setosa': 0,
    'Virginica': 1,
    'Versicolor': 2
})

from sklearn import linear_model
from sklearn.model_selection import train_test_split

x = df[['sepal_length_cm', 'sepal_width_cm', 'petal_length_cm', 'petal_width_cm']]
y = df['class']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train, y_train)

print(model.predict([[4, 3, 2, 0.5]]))

print(model.coef_)
print(model.intercept_)

y_pred = model.predict(x_test)
```

```python
import numpy as np
import pandas as pd

df = pd.read_csv('iris-data-clean.csv')
df['class'] = df['class'].replace({
    'Setosa': 0,
    'Virginica': 1,
    'Versicolor': 2
})

from sklearn import linear_model
from sklearn.model_selection import train_test_split

x = df[['sepal_length_cm', 'sepal_width_cm']]
y = df['class']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)

from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_train, y_train)

print(model.predict([[4, 3]]))

print(model.coef_)
print(model.intercept_)

y_pred = model.predict(x_test)
```
